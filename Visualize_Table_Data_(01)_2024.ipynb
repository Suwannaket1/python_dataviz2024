```
import pandas as pd
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

# Parallel Coordinates
เอาคอลัมน์ที่สนใจมาเรียงต่อกันแล้ว plot ค่า

```
pd.plotting.parallel_coordinates(df, 'Name');
```

## ปรับ parameter: color

```
#เปลี่ยนสี html color เป็นเลขฐาน 16
pd.plotting.parallel_coordinates(df, 'Name', color=('#99CCFF', '#8470FF', '#FFD700'))
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=((0.5, 0.8, 1), (0, 0.7, 0), (0.6, 0, 1))) #rgb value (r,g,b) | r,g,b -> [0,1] / 1 คือสว่างสุด , 0 คือเข้มสุด
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=('y', 'm', 'c')) #ตัวย่อของสี R G B W(white) C M Y K
```

## ปรับ parameter: colormap

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='PuRd')
```

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='tab10')
```

# เริ่ม Matplotlib

```
from matplotlib import pyplot as plt
```

### แปลง data จาก pandas dataframe เป็น numpy array

```
np_data = df.iloc[:,:-1].to_numpy()
np_data
```

# Visualize array data ด้วย pixel

```
import matplotlib
```

```
matplotlib.rcParams['figure.figsize']=[15,15] #กำหนดขนาดของ figure #ค่ามากเป็นสีสว่าง ค่าน้อยเป็นสีเข้ม

plt.imshow(np_data[:50,:],interpolation='nearest')
```

```
plt.imshow(np_data[50:100,:],interpolation='nearest')
```

# การใช้ subplot

```
plt.subplot(1,3,1)  # subplot(มีกี่แถว, มีกี่หลัก, เป็นตัวที่เท่าไหร่) #ค่ามากเป็นสีสว่าง ค่าน้อยเป็นสีเข้ม
plt.imshow(np_data[:50,:],interpolation='nearest',cmap='magma')

plt.subplot(1,3,2)
plt.imshow(np_data[50:100,:],interpolation='nearest',cmap='magma')

plt.subplot(1,3,3)
plt.imshow(np_data[100:150,:],interpolation='nearest',cmap='magma')
```

```
plt.imshow(np_data[:10,:],interpolation='nearest',cmap='magma',alpha = 0.5 ) #alpha เป็นความเข้มโปร่ง ค่าน้อยโปร่งใส [0,1]
```
