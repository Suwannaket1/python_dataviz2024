#โจทย์ข้อที่ 1

ให้หาความแตกต่างของ ค่าเฉลี่ย ของข้อมูล Chlo_a ทุกๆ station ระหว่างปี 2018 และ ปี 2019

```python
# แยกปีจากคอลัมน์ month
Tidy_df['year'] = Tidy_df['month'].apply(lambda x: x.split()[1]) # ['July', '2018']
```

```python
# คำนวณค่าเฉลี่ย Chlo_a สำหรับแต่ละ station ระหว่างปี 2018 และปี 2019
avg_2018 = Tidy_df[Tidy_df['year'] == '2018'].groupby('station')['Chlo_a'].mean()
avg_2019 = Tidy_df[Tidy_df['year'] == '2019'].groupby('station')['Chlo_a'].mean()
```

```python
# ผสมค่าเฉลี่ยเข้าด้วยกันใน DataFrame เดียว
avg_diff = pd.DataFrame({'2018': avg_2018, '2019': avg_2019})
```

```python
# คำนวณความแตกต่างระหว่างปี 2018 และ 2019
avg_diff['difference'] = avg_diff['2019'] - avg_diff['2018']
```

```python
# แสดงผลลัพธ์
avg_diff
```

#โจทย์ข้อที่ 2

ให้สร้างตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน

```python
# แยกเดือนและปีออกจากคอลัมน์ month
Tidy_df['year'] = Tidy_df['month'].apply(lambda x: x.split()[1])
Tidy_df['month_only'] = Tidy_df['month'].apply(lambda x: x.split()[0])
```

```python
# กำหนดลำดับของเดือน
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June',
               'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
Tidy_df['month_only'] = pd.Categorical(Tidy_df['month_only'], categories=month_order, ordered=True)
```

```python
# คำนวณค่าเฉลี่ย Chlo_b สำหรับแต่ละ station ในแต่ละเดือน
monthly_avg = Tidy_df.groupby(['station', 'month_only'])['Chlo_b'].mean().reset_index()
```

```python
# แปลง DataFrame ให้มีรูปแบบที่ง่ายต่อการอ่าน
monthly_avg_pivot = monthly_avg.pivot(index='station', columns='month_only', values='Chlo_b')
```

```python
# กรองคอลัมน์ที่มีค่า null ออก
monthly_avg_pivot = monthly_avg_pivot.dropna(axis=1, how='all')
```
