```python
#ดู data ด้วย Box Plot และ Histogram
import matplotlib.pyplot as plt
import seaborn as sns

def plot_boxplot_with_quartiles(data, column, title):
    plt.figure(figsize=(10, 6))
    ax = sns.boxplot(x=data[column])

    # คำนวณ Quartiles
    Q1 = data[column].quantile(0.25)
    Q2 = data[column].quantile(0.50)
    Q3 = data[column].quantile(0.75)

    # แสดง Quartiles บนกราฟ
    plt.axvline(Q1, color='r', linestyle='--', label='Q1 (25th Percentile)')
    plt.axvline(Q2, color='g', linestyle='--', label='Q2 (Median / 50th Percentile)')
    plt.axvline(Q3, color='b', linestyle='--', label='Q3 (75th Percentile)')

    plt.title(title)
    plt.xlabel(column)
    plt.legend()
    plt.show()
```

#จุดที่น่าสนใจของข้อมูล จุดที่ 1

```python
# สร้าง Box Plot สำหรับคอลัมน์ 'Total Likes'
plot_boxplot_with_quartiles(Doctor_df, 'Total Likes', 'Box Plot of Total Likes')
```

```python
# สร้าง Box Plot สำหรับคอลัมน์ 'Number of replies'
plot_boxplot_with_quartiles(Doctor_df, 'Number of replies', 'Box Plot of Number of replies')
```

```python
# สร้าง Histogram สำหรับคอลัมน์ 'Total Likes'
plt.figure(figsize=(12, 6))
plt.hist(Doctor_df['Total Likes'], bins=50, color='blue', alpha=0.7)
plt.title('Histogram of Total Likes')
plt.xlabel('Total Likes')
plt.ylabel('Quantity')
plt.show()
```

```python
#ลิสต์ของชนิดมะเร็งที่พบบ่อยในคนไทย
cancer_types = ['มะเร็งตับ', 'มะเร็งท่อน้ำดี', 'มะเร็งปอด', 'มะเร็งลำไส้ใหญ่', 'มะเร็งเต้านม', 'มะเร็งปากมดลูก']

import numpy as np

# ฟังก์ชันตรวจสอบว่าคอลัมน์ comments มีคำจากลิสต์หรือไม่
def contains_cancer_type(comment):
    if pd.isna(comment):  # ตรวจสอบค่าที่เป็น NaN
        return np.nan
    found_types = [cancer_type for cancer_type in cancer_types if cancer_type in comment]
    if found_types:
        return ', '.join(found_types)
    return np.nan

# สร้างคอลัมน์ใหม่ที่บรรจุชนิดของมะเร็งที่พบในแต่ละ comments ถ้าไม่มีพบชนิดมะเร็งตามลิสต์จะให้ค่าเป็น null
Doctor_df['Cancer Types Found'] = Doctor_df['comments'].apply(contains_cancer_type)
Doctor_df
```

```python
#เอาเฉพาะ Row ที่เป็นชนิดมะเร็งที่พบบ่อยในคนไทย
Doctor_df.dropna(subset=['Cancer Types Found'], inplace=True)
Doctor_df
```
