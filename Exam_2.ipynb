```python
CrimeSceneReport_df.info()
```

```python
# แปลง Dtype คอลัมน์ date เป็น date
CrimeSceneReport_df['date'] = pd.to_datetime(CrimeSceneReport_df['date'], format='%Y%m%d')

# ปรับรูปแบบวันที่ให้เป็น DD-MM-YYYY
CrimeSceneReport_df['date'] = CrimeSceneReport_df['date'].dt.strftime('%d %m %Y')

CrimeSceneReport_df
```

```python
#ปรับเดือนให้เป็นเดือนภาษาไทย
thai_month = {"01": "มกราคม", "02": "กุมภาพันธ์", "03": "มีนาคม", "04": "เมษายน",
              "05": "พฤษภาคม", "06": "มิถุนายน", "07": "กรกฎาคม", "08": "สิงหาคม",
              "09": "กันยายน", "10": "ตุลาคม", "11": "พฤศจิกายน", "12": "ธันวาคม"}

CrimeSceneReport_df['date'] = CrimeSceneReport_df['date'].apply(lambda x: x.split()[0] + " " + thai_month[x.split()[1]] + " " + x.split()[2])

CrimeSceneReport_df
```

```python
# กรองข้อมูลคดีฆาตรกรรมในวันที่ 15 มกราคม 2018 ที่เมือง SQL City
Crime_df = CrimeSceneReport_df[(CrimeSceneReport_df['type'] == 'murder') &
                               (CrimeSceneReport_df['date'] == '15 มกราคม 2018') &
                               (CrimeSceneReport_df['city'] == 'SQL City')]
Crime_df
```
