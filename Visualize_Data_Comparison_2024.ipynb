```python
import pandas as pd
import os
```

```python
path = '/content/drive/MyDrive/data_viz_2024_DATA'
```

```python
movie_df = pd.read_csv(os.path.join(path,'IMDb_Movies_India.csv'), encoding='latin-1')
movie_df
```

```python
movie_df = movie_df.dropna(subset=['Rating', 'Duration', 'Genre'])
movie_df = movie_df[['Rating', 'Duration', 'Genre']]
movie_df
```

```python
movie_df['Duration'] = movie_df['Duration'].astype(str).str.replace(' min', '') # Convert 'Duration' column to string type
movie_df['Duration'] = movie_df['Duration'].astype(int)
movie_df
```

```python
# prompt: แยกตารางออกเป็น 5 ตารางสำหรับ genre: drama, action, romance, horror, musical

# Create separate DataFrames for each genre
drama_df = movie_df[movie_df['Genre'] == 'Drama']
action_df = movie_df[movie_df['Genre'] == 'Action']
romance_df = movie_df[movie_df['Genre'] == 'Romance']
horror_df = movie_df[movie_df['Genre'] == 'Horror']
musical_df = movie_df[movie_df['Genre'] == 'Musical']
```

```python
all_table = pd.concat([drama_df,action_df,romance_df,horror_df,musical_df])
all_table
```

```python
# prompt: แบ่ง duration ออกเป็น Short Medium Long โดยให้ short น้อยกว่า 60 min Medium 60-90 min ที่เหลือมากกว่า 90 min ให้เป็น Long

def categorize_duration(duration):
  if duration < 60:
    return 'Short'
  elif 60 <= duration <= 90:
    return 'Medium'
  else:
    return 'Long'

all_table['Duration Category'] = all_table['Duration'].apply(categorize_duration)
```

```python
all_table.groupby(['Genre','Duration Category'])['Rating'].mean()
```

```python
india_bar_data = all_table.groupby(['Genre','Duration Category'])['Rating'].mean().unstack()
```

```python
ax = india_bar_data.plot(kind='bar', figsize=(8, 6), width=0.8)

# Add labels and title
plt.xlabel('Region')
plt.ylabel('Rating')
plt.title('Indian movies, separated into groups by Genre and comparing Ratings based on Duration.')

# Move the legend outside the plot
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Adjust the plot size
plt.gcf().subplots_adjust(right=0.8)

# Show the plot
plt.show()
```
